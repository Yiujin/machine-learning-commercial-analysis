```python
# columns들의 리스트
header = list(data)

# standardScaler 함수 사용하여 data 스케일링, 에러는 한글때문임, 한글 외에 모든 데이터 스케일링 됌
def standardScaler(header_list):
    over_array = st_data[header_list]
    a = pd.DataFrame(over_array)
    a["idx"] = np.arange(31547)
    train_data = a.values
    
    from sklearn.preprocessing import StandardScaler
    standardScaler = StandardScaler()
    print(standardScaler.fit(train_data))
    train_data_standardScaled = standardScaler.transform(train_data)
    st_df = pd.DataFrame(train_data_standardScaled, columns=['standard_over','B'])
    del st_df['B']
    st_data[header_list] = st_df

# MinMaxScaler 함수 사용하여 data 스케일링, 에러는 한글때문임, 한글 외에 모든 데이터 스케일링 됌
def minMaxScaler(header_list):
    over_array = mi_data[header_list]
    a = pd.DataFrame(over_array)
    a["idx"] = np.arange(31547)
    train_data = a.values
    
    from sklearn.preprocessing import MinMaxScaler
    minMaxScaler = MinMaxScaler()
    print(minMaxScaler.fit(train_data))
    train_data_minMaxScaled = minMaxScaler.transform(train_data)
    mi_df = pd.DataFrame(train_data_minMaxScaled, columns=['minMax_over','B'])
    del mi_df['B']
    mi_data[header_list] = mi_df
    
# maxAbsScaler 함수 사용하여 data 스케일링, 에러는 한글때문임, 한글 외에 모든 데이터 스케일링 됌
def maxAbsScaler(header_list):
    over_array = ma_data[header_list]
    a = pd.DataFrame(over_array)
    a["idx"] = np.arange(31547)
    train_data = a.values
    
    from sklearn.preprocessing import MaxAbsScaler
    maxAbsScaler = MaxAbsScaler()
    print(maxAbsScaler.fit(train_data))
    train_data_maxAbsScaled = maxAbsScaler.transform(train_data)
    ma_df = pd.DataFrame(train_data_maxAbsScaled, columns=['maxAbs_over','B'])
    del ma_df['B']
    ma_data[header_list] = ma_df
    
# robustScaler 함수 사용하여 data 스케일링, 에러는 한글때문임, 한글 외에 모든 데이터 스케일링 됌
def robustScaler(header_list):
    over_array = ro_data[header_list]
    a = pd.DataFrame(over_array)
    a["idx"] = np.arange(31547)
    train_data = a.values
    
    from sklearn.preprocessing import RobustScaler
    robustScaler = RobustScaler()
    print(robustScaler.fit(train_data))
    train_data_robustScaled = robustScaler.transform(train_data)
    ro_df = pd.DataFrame(train_data_robustScaled, columns=['robust_over','B'])
    del ro_df['B']
    ro_data[header_list] = ro_df

# 기존 data 내에 columns들을 반복문을 통해 모두 스케일링
for a in header:
    if(a == 'business_month_avg' or  a == "10's_sales_amount" or a =='c_month_sales_amount' or a == 'total_moving_people' or a == 'act_jipyo_value' or a == 'growth_jipyo_value' or a == 'safety_jipyo_value'):
        standardScaler("{}".format(a))
        minMaxScaler("{}".format(a))
        maxAbsScaler("{}".format(a))
        robustScaler("{}".format(a))
mi_data
```

```python
import chart_studio.plotly as py
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode()
import chart_studio as py
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import StandardScaler

data=pd.read_csv("plus_living.csv")
del data['Unnamed: 0']
del data['cm_code_name']
test=pd.DataFrame(data,columns=["10's_sales_rate","business_month_avg","20's_sales_rate","store_number","cm_code"])
a = pd.DataFrame(test,columns=["10's_sales_rate"])
a["idx"] = np.arange(31547)
train_data = a.values
    
#1
RobustScaler = RobustScaler()
print(RobustScaler.fit(train_data))
train_data_minMaxScaled = RobustScaler.transform(train_data)
mi_df = pd.DataFrame(train_data_minMaxScaled, columns=['minMax_over','B'])
del mi_df['B']
test["10's_sales_rate_after"] = mi_df
#2
a = pd.DataFrame(test,columns=["business_month_avg"])
a["idx"] = np.arange(31547)
train_data = a.values
train_data_minMaxScaled = RobustScaler.transform(train_data)
mi_df = pd.DataFrame(train_data_minMaxScaled, columns=['minMax_over','B'])
del mi_df['B']
test["business_month_avg_after"] = mi_df
#3
a = pd.DataFrame(test,columns=["20's_sales_rate"])
a["idx"] = np.arange(31547)
train_data = a.values
train_data_minMaxScaled = RobustScaler.transform(train_data)
mi_df = pd.DataFrame(train_data_minMaxScaled, columns=['minMax_over','B'])
del mi_df['B']
test["20's_sales_rate_after"] = mi_df
#4
a = pd.DataFrame(test,columns=["store_number"])
a["idx"] = np.arange(31547)
train_data = a.values
train_data_minMaxScaled = RobustScaler.transform(train_data)
mi_df = pd.DataFrame(train_data_minMaxScaled, columns=['minMax_over','B'])
del mi_df['B']
test["store_number_after"] = mi_df
#clustering
test_kmeans=pd.DataFrame(test,columns=["10's_sales_rate_after","business_month_avg_after","20's_sales_rate_after","store_number_after","cm_code"])

data_points=test_kmeans.values
kmeans=KMeans(n_clusters=3).fit(data_points)
kmeans.labels_
test_kmeans['cluster_id']=kmeans.labels_
new=pd.merge(test,test_kmeans,on=["10's_sales_rate_after","business_month_avg_after","20's_sales_rate_after","store_number_after","cm_code"])
new=pd.merge(data,new,on=["business_month_avg","20's_sales_rate","10's_sales_rate","store_number","cm_code"])
new=new.drop_duplicates()
del new["10's_sales_rate_after"]
del new["business_month_avg_after"]
del new["20's_sales_rate_after"]
del new["store_number_after"]
test0=new[new['cluster_id']==0]
test1=new[new['cluster_id']==1]
test2=new[new['cluster_id']==2]

sns.lmplot('total_moving_people','cm_code',data=new,fit_reg=False,
           scatter_kws={"s":0.1},hue="cluster_id")
plt.title("example")
```
